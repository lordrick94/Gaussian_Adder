```python
import numpy as np
from matplotlib import pyplot as plt
#%matplotlib notebook
#photutils
```

```python
t = 100
mu = t/2
b_x = 0.02
b_y = 0.03
x,y = np.meshgrid(np.linspace(0,t,t),np.linspace(0,t,t))
Amp = 1e-2
g = Amp*np.exp(-(((x-mu)**2)/2*b_x**2+((y-mu)**2)/2*b_y**2))
#g.shape
plt.imshow(g)
plt.colorbar()
```

```python
t = 10
r = 32
mu = t/2
mu_z = r/2
b_x = 0.02
b_y = 0.03
b_z = 0.03
x,y,z = np.meshgrid(np.linspace(0,t,t),np.linspace(0,t,t),np.linspace(0,r,r))
Amp = 1e-2
g = Amp*np.exp(-(((x-mu)**2)/2*b_x**2+((y-mu)**2)/2*b_y**2)+((z-mu_z)**2/2*b_z**2))
g = np.transpose(g,(2,0,1))
print(g.shape)
print(g)
```
